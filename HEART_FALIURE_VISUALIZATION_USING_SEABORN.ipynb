```python
sns.barplot(x=df["DEATH_EVENT"].value_counts().index,y=df["DEATH_EVENT"].value_counts())
plt.title("DEATH_EVENT")
```

```python
sns.barplot(x=df["diabetes"].value_counts().index,y=df["diabetes"].value_counts())
plt.title("diabetes")
```

```python
sns.barplot(x=df["high_blood_pressure"].value_counts().index,y=df["high_blood_pressure"].value_counts(normalize=True))
plt.title("BLOOD PRESSURE")
```

```python
sns.barplot(x=df["sex"].value_counts().index,y=df["sex"].value_counts(normalize=True))
plt.title("GENDER")
```

```python
sns.barplot(x=df["smoking"].value_counts().index,y=df["smoking"].value_counts(normalize=True))
plt.title("smoking")
```

```python
sns.histplot(x=df["age"],bins=50)
```

```python
sns.barplot(x=df["serum_sodium"],y=df['age'])
plt.tight_layout()
```

```python

plt.violinplot(dataset=df["platelets"].dropna(), showmeans=True, showmedians=False)
plt.title("Platelets Distribution")
plt.show()
```
